### Modelo de limpieza de datos

```python
def limpiar_datos(df):

    # Calcular el porcentaje de valores faltantes por columna
    porcentaje_perdidos = df.isnull().sum() / len(df)

    # Obtener las columnas que tienen m치s del 5% de datos perdidos
    columnas_a_eliminar = porcentaje_perdidos[porcentaje_perdidos > 0.05].index

    # Eliminar las columnas con m치s del 5% de datos perdidos
    df = df.drop(columns=columnas_a_eliminar)

    # Recorrer las columnas restantes y completar los valores faltantes
    for columna in df.columns:
        if df[columna].dtype == 'object':
            # Completar con la moda para columnas de tipo objeto
            moda = df[columna].mode().iloc[0]
            df[columna].fillna(moda, inplace=True)
        else:
            # Completar con la mediana para columnas num칠ricas
            mediana = ld_df[columna].median()
            df[columna].fillna(mediana, inplace=True)

    return df
```
