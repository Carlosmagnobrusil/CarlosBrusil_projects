```python
engine = create_engine('sqlite:///schools.db', echo=True) # creo mi ingenieria para sql
conn = engine.connect() #conecto mi ingenieria 
```

```python
sqlite_table = "school_det"
df_school.to_sql(sqlite_table, conn, if_exists='fail') #añado mi  base al l ingenieria de SQL 
```

```python
sql = """
select EstablishmentName, SchoolLevel, Website
from school_det
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName
from school_det
where DistrictHQDistanceinKm > 100

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_det
where SchoolLevel = 'JHS.'

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_det
where SchoolLevel = 'JHS'

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_det
where EstablishmentName like '% JUNIOR HIGH %' 

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_det
where EstablishmentName like 'C%' 
order by EstablishmentName
limit 10

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_det
where LocatedInRuralOrUrban = 'Urban Area' 
order by EstablishmentName desc


"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName as name
from school_det
where LocatedInRuralOrUrban = 'Urban Area' 
order by EstablishmentName desc


"""

pd.read_sql_query(sql, engine)
```
