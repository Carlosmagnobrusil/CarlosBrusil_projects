```python
engine = create_engine('sqlite:///schools.db', echo=True) # creo mi ingenieria para sql
conn = engine.connect() #conecto mi ingenieria 
```

```python
sqlite_table = "school_deti"
school.to_sql(sqlite_table, conn, if_exists='fail')

sqlite_table2 = "school_facility_details"
facility.to_sql(sqlite_table2, conn, if_exists='fail')
```

```python
sql = """
select EstablishmentName, SchoolLevel, Website
from school_deti
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName
from school_deti
where DistrictHQDistanceinKm > 100

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_deti
where SchoolLevel = 'JHS.'

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_deti
where SchoolLevel = 'JHS'

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_deti
where EstablishmentName like '% JUNIOR HIGH %' 

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_deti
where EstablishmentName like 'C%' 
order by EstablishmentName
limit 10

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName,
SchoolLevel,
DistrictHQDistanceinKm
from school_deti
where LocatedInRuralOrUrban = 'Urban Area' 
order by EstablishmentName desc


"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select EstablishmentName as name
from school_deti
where LocatedInRuralOrUrban = 'Urban Area' 
order by EstablishmentName desc


"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select SchoolLevel,
avg (DistrictHQDistanceinKm) as avg_bac_distancia
from school_deti
Group by SchoolLevel

"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select *,
avg(DistrictHQDistanceinKm) as avg_bac_distancia
from school_deti
where DistrictHQDistanceinKm < 10
Group by SchoolLevel


"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select *
from school_deti
inner join  school_facility_details on school_deti.EstablishmentName = school_facility_details.EstablishmentName
where school_deti.SchoolLevel = 'PS' and school_facility_details.IsLibraryAvailable = 'Not Available'
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select *
from school_deti
inner join  school_facility_details on school_deti.EstablishmentName = school_facility_details.EstablishmentName
where school_deti.SchoolLevel = 'PS' and school_facility_details.IsLibraryAvailable = 'Not Available'
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select *
from school_deti
inner join  school_facility_details on school_deti.EstablishmentName = school_facility_details.EstablishmentName
where school_deti.SchoolLevel = 'PS' and school_facility_details.IsLibraryAvailable = 'Not Available'
Order by school_deti.EstablishmentName
"""

pd.read_sql_query(sql, engine)
```
