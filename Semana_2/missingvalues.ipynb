```python
import pandas as pd #importo la libreria panda 
bikes =  '/content/drive/MyDrive/Coding_Dojo/Semana 2/Pandas Data science/Data /RailsToTrails_National_Count_Data_week.xlsx' #Nombro a la base
bikes_df = pd.read_excel(bikes) #le doy un nombre a la trama de datos  ene ste caso el sugerido en la tarea 
bikes_df.head()
```

```python
bikes_df.info()
```

### Cambiar nombre de columnas 

```python
bikes_df = bikes_df.rename(columns = {'Unnamed: 6':'percent_change'}) #utilizo el metodo panda de rename y ledigo el valor que deseo cambiar al poner colums
bikes_df.head()
```

### Eliminar columnas 

```python
bikes_df.drop(bikes_df.columns[bikes_df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True) #en esta linea le pido al programa que se eliminen las columnas
#que contengan en su nombre unnamed 
bikes_df.head()
```

### Llenar columnas con datos vacios

```python
bikes_df.head()
```

```python
bikes_df.loc[0:51 , "2021 Counts"].fillna(0, inplace = True) #lleno el valor de la columna com puntos ya que no hay ningun otro valor 
bikes_df["2021 Counts"].head()
```
